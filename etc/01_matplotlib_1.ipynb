```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### 기본 그래프

```python
# 기본 그래프 그리기
data = np.random.randint(10, size=10)
data
```

```python
plt.plot(data)
plt.show()
```

```python
# x값 추가
x = np.arange(0, 200, 20)
x
```

```python
plt.plot(x, data)
plt.show()
```

```python
plt.scatter(x, data)
plt.show()
```

### 스타일 설정
- 색상, 마커, 라인 스타일
- 커스텀 스타일 : 선색상, 선굵기, 마커선굵기, 마커색상 ...

색상
- blue : b
- green : g
- red : r
- cyan : c
- magenta : m
- yellow : y
- black : k
- white : w

마커
- . : point marker
- , : pixel marker
- o : circle marker
- v : triangle_down marker
- ^ : triangle_up marker
- < : triangle_left marker
- `>` : triangle_right marker
- 1 : tri_down marker
- 2 : tri_up marker
- 3 : tri_left marker
- 4 : tri_right marker
- s : square marker
- p : pentagon marker
- `*` : star marker
- h : hexagon1 marker
- H : hexagon2 marker
- `+` : plus marker
- x : x marker
- D : diamond marker
- d : thin_diamond marker

라인 스타일
- `-` : solid line style
- `--` : dashed line style
- `-.` : dash-dot line style
- `:` : dotted line style

커스텀 스타일
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 선 색깔
- mew : 마커 선 굵기
- mfc : 마커 내부 색깔

```python
plt.plot(data, "go-.")
plt.show()
```

```python
plt.plot(data, c="b", ls="--", marker="o", lw=2, ms="10", mec="r", mew="3", mfc="b")
plt.show()
```

```python
# limit 설정 : x축과 y축의 범위를 설정할수 있습니다.
plt.xlim(-5, 20)
plt.ylim(-3, 10)
plt.plot(data)
plt.show()
```

```python
# tick 설정
plt.plot(data)
plt.xticks(range(0, 10))
plt.yticks([1, 2])
plt.show()
```

```python
# 그리드 추가
plt.grid(True)
plt.plot(data)
plt.show()
```

```python
# 멀티라인 그리기
datas = [np.random.randint(5, size=5) for _ in range(3)]
datas[0], datas[1], datas[2]
```

```python
# plot 함수를 한번 사용
x = range(5)
plt.plot(x, datas[0], "ro--", x, datas[1], "b*-.", x, datas[2], "g^-")
plt.show()
```

```python
# plot 함수를 여러번 사용해서 여러개의 데이터 그리기
plt.plot(x, datas[0])
plt.plot(x, datas[1])
plt.plot(x, datas[2])
plt.show()
```

legend 범례
- 위치 : 0 ~ 10
- best : 0
- ur : 1
- ul : 2
- ll : 3
- lr : 4
- r : 5
- cl : 6
- cr : 7
- lc : 8
- uc : 9
- center : 10

```python
datas = [np.random.randint(5, size=5) for _ in range(3)]
plt.plot(x, datas[0], label="data0")
plt.plot(x, datas[1], label="data1")
plt.plot(x, datas[2], label="data2")
plt.legend(loc=0)
plt.show()
```

```python
# title 추가
plt.plot(data)
plt.title("Title")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 그래프 크기 조절
plt.figure(figsize=(10, 3)) # (가로, 세로)
plt.plot(data)
plt.show()
```

```python
# 여러개의 그래프 그리기
plt.subplot(2, 1, 1) # 2*1 행렬에서 1번위치에 그리기
plt.plot(datas[0])

plt.subplot(2, 1, 2) # 2*1 행렬에서 2번위치에 그리기
plt.plot(datas[1])

plt.show()
```

```python
datas = [np.random.randint(5, size=5) for _ in range(4)]

for idx in range(4):
    plt.subplot(221 + idx)
    plt.plot(datas[idx])
```
