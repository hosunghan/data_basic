### Matplotlib
- 여러가지 차트 사용
- seaborn
- 데이터 프레임으로 데이터를 받아서 그래프를 그리기

```python
# bar plot
data = np.random.randint(10, size=5)
data
```

```python
plt.bar(range(5), data, width=0.4)
plt.show()
```

```python
plt.bar(range(5), data, yerr=np.random.randint(1, 3, size=5), alpha=0.6)
plt.show()
```

```python
plt.barh(range(5), data, xerr=np.random.randint(1, 3, size=5), alpha=0.6)
plt.show()
```

```python
data1 = np.random.randint(1, 5, size=5)
data2 = np.random.randint(5, 10, size=5)
plt.bar(range(5), data2)
plt.bar(range(5), data1)
plt.show()
```

```python
plt.pie(data, labels=list("ABCDE"), colors=[
        "gold", "red", "blue", "lightgreen", "yellowgreen"], autopct="%1.2f%%")
plt.axis("equal")
plt.show()
```

```python
# scatter plot
x = np.random.rand(200) # 균등분포
y = np.random.rand(200)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randn(200) # 정규분포
y = np.random.randn(200)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(100, size=200) # 균등분포, 정수형
y = np.random.randint(100, size=200)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(10, size=20)
y = np.random.randint(10, size=20)
colors = np.random.rand(20)
area = np.pi * np.arange(10, 50, 2) ** 2
plt.scatter(x, y, c=colors, s=area, alpha=0.5)
plt.show()
```

```python
# histogram
y = np.random.randn(200)
plt.hist(y, bins=10, density=True)
plt.show()
```

### Example
- https://github.com/mwaskom/seaborn-data

```python
# iris
iris = sns.load_dataset("iris")
iris.tail(2)
```

```python
iris["species"].unique()
```

```python
sns.lmplot("sepal_length", "sepal_width", data=iris, hue="species", fit_reg=False)
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
# 여객기 승객수 : heatmap
flights = sns.load_dataset("flights")
flights.tail(2)
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
flights_pivot
```

```python
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.show()
```
